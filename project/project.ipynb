The dataset I'm analyzing is about the environment.
It came from http://stat.cmu.edu/datasets/visualizing.data.zip
I expected to see a connection between the ozone layer and radiation level in different places.

```python
import csv
f = open('environmental.csv' ,'rb')
r = csv.reader(f)
rownum = 0
ozone = []
radiation = []
for row in r:
    if rownum != 0:
        a = map(float, row[0:])
        ozone.append(a[0])
        radiation.append(a[1])
    rownum += 1
print ozone
print radiation
f.close()

```

```python
from pylab import *
```

```python
scatter(ozone, radiation)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('ozone and radiation')
ax.set_xlabel('ozone layer')
ax.set_ylabel('radiation level')
scatter(ozone, radiation)
savefig("environmental_labels_scatter.png")
```

The plot shows the relationship between the ozone layer and radiation level.
As opposed to my expectations, one cannot find a clear connection between the ozone layer and radiation level.
The relationship is not strictly defined.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('ozone and radiation')
ax.set_xlabel('ozone layer')
ax.set_ylabel('radiation level')
bar(ozone, radiation)
savefig("environmental_labels_bar.png")
```

This bar graph shows ozone layers and the radiation levels that match them.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('ozone and radiation')
ax.set_xlabel('ozone layer')
ax.set_ylabel('radiation level')
plot(ozone, radiation)
savefig("environmental_labels_plot.png")
```

As shown in the graph, the ozone and radiation values are very disorganized and don't have a clear connection.
