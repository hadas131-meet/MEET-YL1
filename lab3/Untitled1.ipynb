```python
import csv

f = open('cig.csv' ,'rb')
r = csv.reader(f)
tar = []
nicotine = []
for row in r:
    a=row[0].replace('\t', ' ')
    b=a.split()
    c=map(float,b[1:])
    tar.append(c[0])
    nicotine.append(c[1])
print tar
print nicotine
    
f.close()
```

```python
from pylab import*
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('tar and nicotine')
ax.set_xlabel('tar content')
ax.set_ylabel('nicotine content')
scatter(tar, nicotine)
savefig("cig_labels_scatter.png")
```
